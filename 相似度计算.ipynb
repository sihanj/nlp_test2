```python
import pandas as pd
import numpy as np
file=pd.read_csv('medical.csv')
filenames=list(file.id)
file.info()
```

```python
file=file.drop_duplicates()
```

```python
##相似度计算
import jieba
import jieba.posseg as pseg
import codecs
from gensim import corpora, models, similarities
import os
import os.path
import pandas as pd
stop_words = 'stopword.txt'
stopwords = codecs.open(stop_words,'r',encoding='utf-8').readlines()
stopwords = [w.strip() for w in stopwords]
stop_flag = ['x', 'c', 'u','d', 'p', 't', 'uj', 'm', 'f', 'r']
#停用词性 [标点符号、连词、助词、副词、介词、时语素、‘的’、数词、方位词、代词]
pd.set_option('display.float_format', lambda x: '%.4f' % x) #取消科学计数法
def tokenization(filename):
    result=[]
    text=str(file.seg[file.id==filename])
    jieba.load_userdict("C:/python/medicine_dict.txt")
    words = pseg.cut(text)
    for word, flag in words:
        if flag not in stop_flag and word not in stopwords:
            result.append(word)
    return result

filenames=list(file.id)

corpus = []
for each in filenames:
    corpus.append(tokenization(each))
    
dictionary = corpora.Dictionary(corpus)
doc_vectors = [dictionary.doc2bow(text) for text in corpus]
#TF-IDF
tfidf = models.TfidfModel(doc_vectors)
tfidf_vectors = tfidf[doc_vectors]

d={'ID':pd.Series(filenames)} 
df=pd.DataFrame(d)
for f_id in filenames:
    query = tokenization(f_id)
    query_bow = dictionary.doc2bow(query)
    index = similarities.MatrixSimilarity(tfidf_vectors)
    sims = index[query_bow]
#print(list(enumerate(sims)))
    lst=[]
    for i in enumerate(sims):
        lst.append(i[1])
        
    df[f_id]=lst
df
```

```python
import pandas as pd
df=pd.read_csv('med_cor_mat.csv')
filenames=list(df.ID)
```

```python
fid=[]
data=[]
val=[]
for a in filenames:
    cond=df.ID[df[str(a)]>0.1]
    cond_l=list(cond)
    value=round(df[str(a)][df[str(a)]>0.1],4)
    value_c=list(value)
    if len(cond_l)>1:
        fid.append(a)
        data.append(cond_l)
        val.append(value_c)
```

```python
sim_df=pd.DataFrame(fid,columns=['fid'])
sim_df['data']=data
sim_df['score']=val
sim_df.to_excel('med_sim_df_0.1.xlsx')
```
